## 1.1 Dealing with categorical data

We need to convert the categorical data into a numerical representation in order to experiment with a logistic regression. For now, let's do one hot encoding, the categories have quite a few levels such that using dummy variables may make the model too large (we can also considering encoding with hash function for better efficiency).

```python
#Let's get the relevant columns and split data for training and testing
from sklearn.model_selection import train_test_split

X_train = nets_data.loc[:, nets_data.columns != 'sold'] #Select everything but sold
X_train = X_train.drop('profit', 1) #drop profit for now, leave margin in as it gives us information about cost
Y_train = nets_data['sold'] #Select only sold 

X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=0.33) #Split 66% vs 33%
```

```python
#Use preprocessing to encode the data. Used it for dates as well, not sure if thats the best method???
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression

categorical_features = ['channel','event_date', 'invoice_date', 'opponent', 'pc','ticket_type', 'vendor', 'row', 'section']
X_train_ordinal = X_train[categorical_features]
X_test_ordinal = X_test[categorical_features]

#Use LabelEncoder on the categorical features
le = preprocessing.LabelEncoder()
X_train_ordinal = X_train_ordinal.apply(le.fit_transform)
X_test_ordinal = X_test_ordinal.apply(le.fit_transform)
                                  
#Join numerical columns back 
numerical_features = ['begseat', 'cost', 'endseat', 'margin', 'quantity', 'revenue']
X_train_ordinal = X_train_ordinal.join(X_train[numerical_features], how='outer')
X_test_ordinal = X_test_ordinal.join(X_test[numerical_features], how='outer')
```

```python
#Lets try fitting the model and see
l = LogisticRegression()
l.fit(X_train_ordinal, Y_train)
Y_pred = l.predict_proba(X_test_ordinal)
```

```python
from sklearn.metrics import log_loss
log_loss = log_loss(Y_test,Y_pred)

acc = l.score(X_test_ordinal, Y_test)

print(log_loss)
print(acc)
```
